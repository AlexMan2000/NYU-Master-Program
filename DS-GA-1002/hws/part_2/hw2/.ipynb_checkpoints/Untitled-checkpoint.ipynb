```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
data = pd.read_csv("./ANSUR II MALE Public.csv",encoding="gbk")
```

```python
data.columns
```

```python
X = data["Heightin"]
Y = data["Weightlbs"]
```

```python
covXY = np.cov(X,Y)[0][1]
sY = np.std(Y)
sX = np.std(X)
```

```python
beta1 = covXY / (sX ** 2)
beta0 = np.mean(Y) - beta1 * np.mean(X)
residuals = Y - beta0 - beta1 * X
np.cov(X, residuals)
```

```python
plt.scatter(X,Y)
plt.plot(X, beta0 + beta1 * X, c="red")
plt.xlabel("height")
plt.ylabel("weight")
```

```python
predictor = beta0 + beta1 * X
np.var(Y)
```

```python
np.var(predictor) + np.var(residuals)
```

```python
R2 = np.var(predictor) / np.var(Y)
R2
```
