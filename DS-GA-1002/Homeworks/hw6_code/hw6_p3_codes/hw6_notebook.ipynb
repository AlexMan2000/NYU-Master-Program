```python
# Problem 3(b): Simulate the Markov Chain
x0 = np.array([0.9,0.1,0])
T = np.array([[0.7,0,0],[0.2,1,0],[0.1,0,1]])
state_name = ["In Stock", "Sold Locally", "Exported"]


def one_step(state, transition_matrix):
    return np.random.choice(3, 1, p = transition_matrix[:, state].flatten())[0]


def simulate_discrete_markov_chain(start, transition_matrix, step):
    trace = [start]
    for i in range(step):
        trace.append(one_step(start, transition_matrix))
        
    return trace

def plot_trace(state_name, start_prob, transition_matrix, step):
    start_state = np.random.choice(len(state_name), 1, p = start_prob)[0]
    trace = simulate_discrete_markov_chain(start_state, transition_matrix, step)
    print(trace)
    plt.plot(np.arange(0, step + 1), trace, "o--")
    plt.yticks([0,1,2], state_name)
    
    plt.show()
    
plot_trace(state_name, x0, T, 20)
```
