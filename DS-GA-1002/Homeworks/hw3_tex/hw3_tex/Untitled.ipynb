```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def rect_kernel(data, center, width):
    return 1/width if data <= center + width/2 and data >= center - width/2 else 0

def compute_kde(data_points, t):
    return  1 / len(data_points) * sum([rect_kernel(t, data_point, 2) for data_point in data_points])

def compute_prob(data_points, low, high):
    return sum([compute_kde(data_points, t) for t in range(low, high + 1)])
```

```python
data_points = [5,8,5,6]
compute_prob(data_points, 6, 20)
```

```python
compute_prob(data_points, 7, 20)
```

```python
indices = np.linspace(0,20,100)
plt.plot(indices, [compute_kde(data_points, index) for index in indices])
```

```python
def piecewise_function(t):
    if t < -1:
        return 0
    elif -1 <= t <= 0:
        return 0.5 * t + 0.5
    else:
        return 1-0.5*np.exp(-t)

x_axis = np.linspace(-3,4,100)    
    
plt.plot(x_axis,[piecewise_function(t) for t in x_axis])
```
